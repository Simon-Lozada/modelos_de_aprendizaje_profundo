Las máquinas de Boltzmann restringidas son un tipo de red neuronal simple pero no trivial. Los RBM son interesantes porque se pueden usar con una red neuronal más grande para realizar el preentrenamiento : encontrar pesos "buenos predeterminados" para una red neuronal sin tener que hacer un trabajo de retropropogación más costoso, levantando pesos de una red más simple que entrenamos de antemano que enseñamos a reconstruir aproximadamente los datos de entrada. El entrenamiento previo ha sido históricamente importante en el desarrollo de redes neuronales porque permite entrenar grandes redes neuronales de manera más rápida y confiable de lo que sería posible de otra manera. Veremos cómo lo hacen en este cuaderno.

Tenga en cuenta que el preentrenamiento ha caído un poco en desgracia en la vanguardia porque ha sido algo obviado por mejores algoritmos de optimización . Además, los RBM han sido algo desplazados por codificadores automáticos más complicados pero más robustos (tema para el futuro).

Los RBM son una red poco profunda con solo dos capas: una capa de entrada y una capa oculta. Los nodos de la capa de entrada están completamente conectados a los nodos de la capa oculta ... y los nodos de las capas ocultas, a su vez, están completamente conectados a los nodos de la capa de entrada. Por tanto, la señal rebota entre los nodos de los nodos de la capa de entrada y la capa oculta y viceversa.

El RBM recicla pesos. El peso de la conexión hacia adelante entre un nodo de entrada y un nodo oculto es el mismo que el peso de la hacia atrás conexión entre el nodo oculto y el nodo de entrada.

El RBM aprende en cada pase. En el pase hacia adelante (proyección), la reconstrucción se trata como el objetivo y el descenso de gradiente se usa para ajustar los pesos en la capa de entrada. En el paso hacia atrás, el conjunto de datos original se trata como el objetivo y el descenso de gradiente se usa para ajustar los pesos en la capa oculta. Dado que los pesos hacia atrás y hacia adelante son los mismos, cada pasada ajusta simultáneamente los pesos hacia adelante y hacia atrás.

Estamos gestionando efectivamente dos distribuciones. Una distribución es una función de los datos de entrada, f ( X ) = X p
. La otra distribución es una función de esa salida, g ( X p ) = X q . Haciendo los pesos que controlan f ( X ) y g ( X ) lo mismo, y al realizar la retropropogración en cada paso a través de la red, estamos logrando f ( X p ) y g ( X p ) convergen aproximadamente en la misma distribución. En otras palabras, f ( X ) ≈ g - 1 ( X )

y viceversa: las funciones de avance y retroceso se vuelven aproximadamente inversas entre sí.

Para medir la distancia entre su distribución de probabilidad estimada y la distribución de verdad del terreno de la entrada, los RBM utilizan Kullback Leibler Divergence. KL-Divergence mide las áreas divergentes bajo las curvas de distribución formadas por X p
y X q

. Una divergencia baja significa que las distribuciones son similares, lo que a su vez significa que son aproximaciones más cercanas entre sí.

Hacer todo esto le permite ganar peso y términos de sesgo en sus nodos ocultos que crean una reconstrucción efectiva de su capa de entrada cuando los ejecuta. Si usted tiene norte

capas en la red neuronal que realmente está interesado en construir, y repite este procedimiento para cada capa en esa otra red paso a paso, terminará con buenos pesos que puede establecer en esa red que manipulan los datos de entrada tan poco como posible cuando se pasa a través de la red. Este "valor predeterminado sin operación" es un punto de partida mucho mejor para luego realizar su entrenamiento real, en comparación con comenzar con pesos aleatorios, y acelerará enormemente su entrenamiento.

Dicho de otra manera, la capa oculta en un RBM aprende una representación dispersa de los datos.

Los RBM no suelen utilizar los mismos algoritmos de optimización que utilizan otros tipos de redes neuronales (por ejemplo, SGD, ADAM, etc.). En su lugar, los RBM utilizan un algoritmo diferente, aún dentro de un marco de descenso de gradiente, conocido como "divergencia contrastiva". Este algoritmo es el resultado de una larga búsqueda para encontrar un algoritmo que sea lo suficientemente rápido y lo suficientemente preciso para que las GBR sean útiles en la práctica; Geoff Hinton alega que le llevó 17 años finalmente tropezar con él.

La definición clásica de una GBR es categórica. Los valores de la capa objetivo deben ser senderos binarios (0 o 1) o probabilidades (en el rango de 0 a 1). 

## Demo

Demostraremos cómo crear e interpretar un RBM utilizando el conjunto de datos clásico MNIST.

Las GBR no se implementan en keras, porque como se mencionó en el encabezado han caído en desgracia. Puede implementarlos usted mismo usando caffeo tensorflowo un marco similar de bajo nivel (o copie uno de los fragmentos de código que flotan en línea que hacen precisamente eso), o puede usar el scikit-learnimplementación.

```python
import pandas as pd
import numpy as np

X_train = pd.read_csv('data/train.csv').values[:,1:]  # exclude the target
X_train = (X_train - np.min(X_train, 0)) / (np.max(X_train, 0) + 0.0001)  # rescale to (0, 1)
```



Tenga en cuenta que el sklearnclase para el RBM se llama muy específicamente el BernoulliRBM.

Recuerde que la distribución de Bernoulli es una distribución simple donde un valor puede ser 1 o 0. En el conjunto de datos MNIST, cada píxel está en el rango (0, 255), pero simplificaremos los datos para que sean solo 0 o 1. En ese caso tenemos n × n
Bernoulli distribuyó puntos de datos para trabajar, donde norte es la dimensionalidad de la imagen (y n × n

es su recuento total de píxeles).

No hay ningún RBM que no sea el BernoulliRBM. Entonces el sklearnLa RBM solo se puede usar para pruebas binarias o para probabilidades, como se menciona al principio. 

```python
from sklearn.neural_network import BernoulliRBM
rbm = BernoulliRBM(n_components=100, learning_rate=0.01, random_state=42, verbose=True)
rbm.fit(X_train)
```

Para reconstruir datos usando RBM, use el método de gibbs.

```python


def gen_mnist_image(X):
    return np.rollaxis(np.rollaxis(X[0:200].reshape(20, -1, 28, 28), 0, 2), 1, 3).reshape(-1, 20 * 28)

xx = X_train[:40].copy()
for _ in range(1000):
    for n in range(40):
        xx[n] = rbm.gibbs(xx[n])
        
import matplotlib.pyplot as plt
plt.figure(figsize=(10,20))
plt.imshow(gen_mnist_image(xx), cmap='gray')

```

Para ver la lista de componentes extraídos por el RBm ver componentes_.

```python


plt.figure(figsize=(20, 20))
for i, comp in enumerate(rbm.components_):
    plt.subplot(10, 10, i + 1)
    plt.imshow(comp.reshape((28, 28)), cmap=plt.cm.RdBu,
               interpolation='nearest', vmin=-2.5, vmax=2.5)
    plt.axis('off')

```
